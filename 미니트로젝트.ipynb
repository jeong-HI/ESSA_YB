```python
df = pd.read_csv('https://s3.ap-northeast-2.amazonaws.com/data10902/petition/petition.csv', parse_dates=['start', 'end'])
```

```python
# 1. 어떤 카테고리에서 가장 많은 청원이 있었는지
ct= df.category.value_counts()
ct
# 정치개혁이 가장 많았음
```

```python
# 2. 어떤 달에 가장 청원이 많았는지(start+category)
df['y']=df['start'].dt.year
df['m']=df['start'].dt.month
df.groupby(['y','m']).category.value_counts()

# 2017년 8월에는 육아/교육, 2019년에는 경제 민주화 카테고리의 청원이 많았던것을 확인 할 수 있다.
# 데이터를 더 출력해보면 다른 년도 다른 달에는 어떤 청원이 있었는지 알 수 있다.
```

```python
 # 년도별 청원수(년도없이 달만)
ax= plt.subplots()
ax= sns.countplot('y', data=df)
#2018년이 많다고 나왔지만 2017년은 8월부터 시작, 2019년은 2월에 끝나므로 2018년이 가장 많다고 보기 어려움
```

```python
# 달별 청원수(년도없이 달만)
ax= plt.subplots()
ax= sns.countplot('m', data=df)
# 1월과 11월에 많은 청원이 있던것으로 보임
```

```python
#청원이 많았던 날
start_df = pd.DataFrame(df['start'].value_counts()).reset_index()
start_df.columns = ['start', 'counts']
start_df = start_df.sort_values('counts', ascending=False)
start_df.head()
# 2017년 11월 11일에 청원이 가장 많았다.
```

```python
#투표가 많았던 날, -참고자료 참고
petitions_start = pd.pivot_table(df, index=['start'], aggfunc=np.sum)
votes_df = petitions_start.sort_values(by='votes', ascending=False)
votes_df.loc[petitions_start['votes'] > 350000].head()

# 대체로 2018년에 투표가 가장 많은날이 많은것으로 보임
# 2018년 2월에 청원도 많고 투표도 많은것을 확인할 수 있음
```

```python
#3. 년도별로 어떤 카테고리 청원이 많았는지(start+category)
df['y']=df['start'].dt.year
j=df.groupby(['y']).category.value_counts()
j
# 2017-2019 모두 정치개혁이 1위, 1-3위는 큰 변화가 안보이고 4위부터 변동이 있는것같다.
```

```python
# 1에서 구한 카테고리별 청원수 시각화... 왜 한글이 꺠질까요...
category = pd.DataFrame(df['category'].value_counts()).reset_index()
category.columns = ['category', 'counts']
category
```

```python
ax= plt.subplots()
ax= sns.barplot('category','counts',data=category)
#글씨가 깨지지만 위의 표와 비교해면 13위~, ~2위 이렇게 청원수가 차이나는것을 알 수 있다.
# 달별 청원수(년도없이 달만)
```

```python
#4. 각 년도/달별 어떤 title과 content가 hot했는지
df.groupby(['y','m']).describe(include=np.object)
# 투표와 청원수가 많던 2018년 2월에 큰 이슈는 김보름 선수이야기인것을 알 수 있음
#청원수가 가장 많았던 2017-11월에는 이명박 출국이 큰 이슈임을 알 수 있음
```

```python
#5. 카테고리별로 어떤 title가 hot했는지
df.groupby(['category'])['title'].max()
#그동안 어느 문제가 있었는지 대략적인 파악 가능
```
